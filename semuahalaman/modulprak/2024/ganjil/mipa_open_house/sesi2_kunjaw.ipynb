## Fungsi

Buatlah fungsi `jumlahkan` yang menerima input `n` lalu menghitung

$$1 + 2 + 3 + \dots + n = \text{ ?}$$

```python
def jumlahkan(n):
  hasil = 0
  for i in range(1, n+1):
    hasil = hasil + i
  return hasil
```

```python
print(jumlahkan(3))
```

```python
print(jumlahkan(100))
```

## Fungsi Rekursif

Barisan Fibonacci: 0, 1, 1, 2, 3, 5, 8, ...

$$F_0 = 0, \quad F_1 = 1$$

$$F_n = F_{n-1} + F_{n-2}$$

```python
def fib(n):
  if (n == 0):
    return 0
  elif (n == 1):
    return 1
  else:
    return fib(n-1) + fib(n-2)
```

$$F_{10} = \text{ ?}$$

```python
fib(10)
```
